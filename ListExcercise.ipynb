```python
# Teht채v채 2
import random
thrownDiceNumbers = [0, 0, 0, 0, 0]
for i in range(5):
    thrownDiceNumbers[i] = random.randint(1, 6)
print("Dice numbers:", thrownDiceNumbers)
total = sum(thrownDiceNumbers)
print("Total of the dice throws", total)
highest = max(thrownDiceNumbers)
print("Highest dice throw:", highest)
```
